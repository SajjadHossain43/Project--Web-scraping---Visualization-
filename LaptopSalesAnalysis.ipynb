```python
import matplotlib.pyplot as plt

market_cut = {}
for i, j in brands.to_dict().items():
  x = LaptopSales[LaptopSales['Brand'].apply(lambda x: x.strip() ==  i)]['Range'].value_counts().to_dict()
  market_cut[i] = x

low_market_cut = {}
middle_market_cut = {}
high_market_cut = {}
for i, j in market_cut.items():
  if j.get('low') != None:
    low_market_cut[i] = j.get('low')
  if j.get('middle') != None:
    middle_market_cut[i] = j.get('middle')
  if j.get('high') != None:
    high_market_cut[i] = j.get('high')

low_labels = list(low_market_cut.keys())
low_values = list(low_market_cut.values())
middle_labels = list(middle_market_cut.keys())
middle_values = list(middle_market_cut.values())
high_labels = list(high_market_cut.keys())
high_values = list(high_market_cut.values())

plt.figure(figsize=(8, 8))
plt.pie(low_values, labels=low_labels, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title("Low Market Share Distribution")
plt.figure(figsize=(8, 8))
plt.pie(middle_values, labels=middle_labels, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title("Middle Market Share Distribution")
plt.figure(figsize=(8, 8))
plt.pie(high_values, labels=high_labels, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title("High Market Share Distribution")
```
